```python
#implementation using arrays
class Stack:
  def __init__(self,max):
    self.max=max
    self.top=-1
    self.data=[]
  def push(self,ele):
    self.ele=ele
    if self.top+1>self.max-1:
      print("Stack overflow")
    else:
      self.top=self.top+1
      self.data.insert(self.top,ele)
  def pop(self):
    if self.top<0:
      print("Stack underflow")
    else:
      print(self.data[self.top])
      self.top=self.top-1
      #self.data.pop()
  def display(self):
    if self.top<0:
      print("Stack Underflow")
    else:
      print("The elements of Stack are:-")
      for i in range(self.top,-1,-1):
        print(self.data[i])
  def isempty(self):
    if self.top==-1:
      return True
    else:
      return False
  def isfull(self):
    if self.top>=self.max:
      return True
    else:
      return False
  def peak(self):
    return self.data[self.top]

max=int(input("Enter size of stack:"))
stc=Stack(max)
while True:
  print("\n1.Push\n2.Pop\n3.Display\n4.Exit")
  ch=int(input("Enter your choice:"))
  
  if ch==1:
    ele=eval(input("Enter element to be inserted:"))
    stc.push(ele)
  elif ch==2:
    stc.pop()
  elif ch==3:
    stc.display()
  elif ch==4:
    break
  elif ch>4:
    print("#Enter Valid Input#")
```

```python
class Node:
    def __init__(self, data):
       self.data = data
       self.next = None
 
class Stack:
    def __init__(self):
        self.head = None
 
    def push(self, data):
        if self.head is None:
            self.head = Node(data)
        else:
            new_node = Node(data)
            new_node.next = self.head
            self.head = new_node
 
    def pop(self):
        if self.head is None:
            return None
        else:
            popped = self.head.data
            self.head = self.head.next
            return popped
    
    def display(self):
      temp=self.head
      print("TOP-->",end=" ")
      while temp.next!=None:
        print(temp.data)
        temp=temp.next
      print(temp.data)
      print()
 
a_stack = Stack()
print('1.push')
print('2.pop')
print('3.display')
print('4.quit')
while True:
    ch=int(input("\nEnter your choice: "))
    if ch==1:
        data=eval(input("Enter data: "))
        a_stack.push(data)
    elif ch==2:
        popped = a_stack.pop()
        if popped is None:
            print('Stack is empty.')
        else:
            print('Popped value: ', int(popped))
    elif ch==3:
      a_stack.display()
    elif ch==4:
        break
    else:
      print("## WRONG INPUT ##")
```
